```python
# import necessary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Read prices.csv file

df = pd.read_csv("prices.csv")
df
```

```python
# Country is limited to USA
# Time Frame is set from year 2000 to 2005
# NVDA Ticker is Selected for this dataframe

df1 = df[(df["Country"] == "usa") & (df["Date"].between("2000-01-01", "2005-12-31") & (df["Ticker"] == "NVDA"))]
df1
```

```python
# Set Date column to date format

df1['Date'] = pd.to_datetime(df['Date'], utc = True)
df1 = df1.sort_values(by = 'Date', ascending = True)
df1
```

```python
# Features not used are removed from the dataframe such as Open, High, Low, Close,
# brand name, ticker, industry tag, country, dividends, stock splits, and capital gains

df2 = df1.drop(["Open", "High", "Low", "Close", "Brand_Name", "Ticker", "Industry_Tag", 
                "Country", "Dividends", "Stock Splits", "Capital Gains"], axis = 1)
df2
```

```python
# Change in Volume (at time = t) = Volume (t) - Volume (t-1)

df2["Volume Change"] = df2["Volume"].diff()

# Set Lag = 1 day using shift function

df2["Lagged Volume"] = df2["Volume"].shift(1)

# Change in Lagged Volume (at time = t) = Lagged Volume (t) - Lagged Volume (t-1)

df2["Lagged Volume Change"] = df2["Lagged Volume"].diff()

# Set Date as index for the dataframe

df2 = df2.set_index("Date")

df2
```

```python
# Plotting each of the independent features across time

plot = df2.plot(subplots = True, kind = "line", figsize = (12, 10), legend = True, title = [
    "Volume for NVDA from 2000 to 2005",
    "Volume Change for NVDA from 2000 to 2005",
    "Lagged Volume for NVDA from 2000 to 2005",
    "Lagged Volume Change for NVDA from 2000 to 2005"
])

for p in plot:
    p.xaxis.set_major_locator(mdates.YearLocator())           
    p.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    p.set_xlim(pd.Timestamp("1999-12-31"), pd.Timestamp("2006-01-01")) 
    p.tick_params(axis="x", rotation=0)
    p.tick_params(labelbottom = True)        

plt.tight_layout()
plt.show()
```
